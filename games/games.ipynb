## Предобработка данных

### Названия столбцов

### Пропуски

Обрабатывать пропуски мы будем в соответствии со следующими установками:

1. Пропуски, которые могут быть заполнены исходя из содержания столбца, в котором они появились, мы заполним.


2. Пропуски, которые заполнить нельзя и при этом число содержащих их строк сравнительно невелико (<5% от общего числа строк), мы удалим.


3. Пропуски, которые заполнить нельзя и число содержащих их строк велико, мы оставим без изменений.

Как уже говорилось выше, пропуски встречаются в столбцах `name`, `year_of_release`, `genre`, `critic_score`, `user_score` и `rating`.

Посмотрим, какую долю занимают ячейки с пропусками в каждом столбце:

```python
(
    pd.DataFrame(
        games.isna().mean(),
        columns=['omissed'])
    .style.format({'omissed': '{:.2%}'})
    .background_gradient('coolwarm')
)
```

В столбцах `name`, `year_of_release` и `genre` пропусков совсем мало. Скорее всего, эти пропуски возникли из-за неполноты данных в источниках, по которым собирался датасет, а также из-за мелких технических ошибок (н-р, не распознались иноязычные символы) при его формировании (чем-то другим сложно объяснить то, что две игры в датасете не имеют названия).

Осмысленно заполнить их нельзя, поэтому мы сразу удалим содержащие их строки. Это не скажется негативно на качестве данных.

```python
games.dropna(subset=['name', 'year_of_release', 'genre'], inplace=True)
```

Проверим результат:

```python
games.info()
```

Пропуски в столбцах `name`, `year_of_release` и `genre` удалены.

Со столбцами `critic_score`, `user_score` и `rating` дело обстоит несколько сложнее. Пропусков в них очень много, так что удалять соответствующие строки было бы неправильно. Тут нужен более детальный анализ. Посмотрим, как соотносятся эти пропуски между собой.

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.heatmap(
    games[['critic_score', 'user_score', 'rating']].isna(),
    yticklabels=False,
    cbar=False,
    cmap='YlGnBu'
)
plt.title('Пропуски в столбцах critic_score, user_score и rating', fontsize=14, y=1.02, fontweight='bold');
```

Ряд пропусков в этих столбцах явно взаимосвязаны между собой. Взглянем на первые 15 строк с пропусками в столбце `critic_scores`:

```python
games[games['rating'].isna()].head(15)
```

Сложно выделить какую-то общую черту игр, не имеющих ни оценок критиков и пользователей, ни возрастного рейтинга. Объяснить это только их непопулярностью или «возрастом» нельзя: так, в приведённой таблице есть игры, которые и популярны, и новы (например, Pokemon X/Pokemon Y 2013 года, продано около 15 млн копий по миру).
Посмотрим на игры, по которым часть информации об их рейтингах всё же есть. Может, так мы лучше поймём причину возникновения пропусков.

```python
games[
    ((games['critic_score'].isna()) & ((~games['user_score'].isna()) | (~games['rating'].isna()))) |
    ((games['user_score'].isna()) & ((~games['critic_score'].isna()) | (~games['rating'].isna()))) |
    ((games['rating'].isna()) & ((~games['user_score'].isna()) | (~games['critic_score'].isna())))
].head(20)
```

Понятнее не стало. Какие-либо взаимосвязи между пропусками и другими данными таблицы выделить не удаётся. Тогда остаётся только одно более или менее убедительное объяснение возникновения пропусков: видимо, информация о рейтингах черпалась с небольшого числа профильных сайтов (возможно, даже с одного сайта); если игры на этих сайтах не было, то её рейтинги было взять неоткуда, так что соответствующие ячейки заполнялись пропусками. Наверняка возникали также ситуации, когда игра была, но рейтингов для неё сайт просто не имел, или имел, но не все. Короче говоря, наиболее вероятная причина пропусков — неполнота исходных данных.

 <a id='tbd'></a>
Кстати, в таблице выше оказалось интересное значение `tbd` (To Be Determined — «Ещё не определён»), наличие которого объясняет, почему столбец `user_score` имеет тип `object`. С этим значением мы разберёмся в блоке [«Артефакты»](#artefacts). 

Что делать с пропусками в столбцах рейтингов?

В столбце `rating` их заполнить можно. Дело в том, что пропуски в нём имеют характер ещё одной категории в добавок к уже имеющимся. Таким образом мы не потеряем часть данных при дальнейшем анализе с участием столбца `rating`.

В качестве заполнителя используем значение `ND` (No Data):

```python
games['rating'].fillna(value='ND', inplace=True)
```

Пропуски в столбцах с оценками критиков и пользователей заполнить при желании тоже реально. В качестве заполнителя можно использовать медианные значения по группам, выделенным на основе трёх признаков одновременно: жанра, года выпуска и диапазону проданных копий в мире. Такая медиана дала бы нам приблизительную оценку возможного мнения критиков и пользователей. В то же время это заполнение очень серьёзно изменило бы данные: всё-таки мы бы заполнили 40-50% пропусков в соответствующих столбцах. Из соображений сохранения чистоты данных не будем заполнять эти пропуски. Оставим их как есть.

На этом обработка пропусков завершена.

### Артефакты

<a id='artefacts'></a>
Судя по типам данных и уже просмотренным уникальным значениям столбцов, [единственный артефакт](#tbd) в датафрейме — это значение `tbd` в столбце `user_score`. Его смысл понятен: оно присваивается играм, по которым оценка пользователей ещё не сформирована. Для нас, впрочем, этот нюанс значения не имеет: нам в этом плане достаточно черно-белого мира, в котором оценка либо есть, либо её нет. Поэтому заменим все значения `tbd` на `NaN` и проверим результат:

```python
games['user_score'] = games['user_score'].where(games['user_score'] != 'tbd', np.nan)

print('Количество строк со значением tbd в столбце user_score:', len(games.query('user_score == "tbd"')))
```

Артефакт успешно устранён.

### Дубликаты

Сначала разберёмся с неявными дубликатами. Их можно искать в столбцах `name`, `platform`, `genre` и `rating`. Сразу упростим себе задачу: искать неявные дубликаты в столбце `name` мы не будем, поскольку в нём больше 11 тысяч уникальных значений (а неуникальные возникают в случаях, когда игра подходит сразу для нескольких платформ). Перебирать все 11 тысяч значений — слишком большая трата времени. Поэтому остановимся на посильных задачах и просмотрим уникальные значения в `platform`, `genre` и `rating`:

```python
for column in ['platform', 'genre', 'rating']:
    display(column, games[column].unique())
    print()
```

Неявных дубликатов в этих столбцах нет, за исключением одного: рейтинг `K-A` — это то же самое, что и `E`, только более старая версия. Заменим `K-A` на `E` и проверим результат:

```python
games['rating'] = games['rating'].where(games['rating'] != 'K-A', 'E')

print('Количество игр с рейтингом K-A:', len(games.query('rating == "K-A"')))
```

Посмотрим, есть ли в столбце `rating` значения, которыми маркировано малое количество игр и при этом они близки по смыслу к другим категориям:

```python
games['rating'].value_counts()
```

Рейтинги `EC`, `AO` и `RP` встречаются редко. Их можно заменить на близкие к ним рейтинги `E`, `M` и `RP` соответственно. Сделаем это и проверим результат.

```python
games['rating'].replace({'EC':'E', 'AO': 'M', 'RP': 'ND'}, inplace=True)
games['rating'].value_counts()
```

Замена прошла успешно.

Посмотрим, есть ли явные дубликаты в датафрейме:

```python
games.duplicated().sum()
```

Явных дубликатов нет. На этом обработка дубликатов завершена.

### Типы данных

Посмотрим ещё раз информацию о столбцах датафрейма:

```python
games.info()
```

Типы данных `float` для столбца с годом и `object` для столбца с оценками пользователей неудобны. Заменим первый на `integer`, а второй на `float`:

```python
games['year_of_release'] = games['year_of_release'].astype('int')
games['user_score'] = games['user_score'].astype('float')

print('Тип столбца user_score:', games['user_score'].dtype)
print('Тип столбца year_of_release:', games['year_of_release'].dtype)
```

Обработка типов данных окончена.

В рамках предобработки мы не рассматривали выбросы, поскольку, судя по [общей статистической информации](#describe) о датасете, выбросы хоть и имеются в столбцах с количеством проданных копий, но они не носят аномального характера: максимальное число проданных копий в одном регионе составляет 41 млн и это вполне реальная цифра.

### Продажи по платформам

Проанализируем теперь динамику продаж по платформам. Сначала посмотрим данные за всё время, а затем перейдём к информации, касающейся базового периода. 

```python
platforms = (games.pivot_table(
        index='year_of_release',
        columns='platform',
        values='total_sales', aggfunc='sum')
    .reset_index()
    .rename_axis(None, axis=1)) # сводная таблица по ежегодным продажам платформ

colors_for_platforms = {
    platf : color for platf, color in zip(
        games['platform'].unique(),
        list(mcolors.XKCD_COLORS.values())[:len(games['platform'].unique()):-1])
} # подбор цвета для каждой из платформ


platforms.plot(x='year_of_release',
               kind='bar',
               stacked=True,
               figsize=(15, 12),
               color=colors_for_platforms,
               rot=45)
plt.xlabel('год выпуска')
plt.ylabel('млн проданных копий')
plt.title('Количество проданных копий игр по платформам в 1980–2016 гг.', fontsize = 16, fontweight='bold')

plt.show()
```

По этой диаграмме можно определить примерный срок жизни платформы. Для относительно крупных платформ он составляет от 5 до 10 лет. После этого платформа может остаться на плаву, но её роль в общих продажах будет крайне незначительна.

Судя по всему, этот срок вполне можно использовать и для прогноза: нельзя сказать, что в последние 20 лет обычное время жизни платформы принципиально менялось.

Посмотрим повнимательнее на платформенную структуру продаж в базовом периоде. Тот факт, что статистика за 2016 неполная, вряд ли серьёзно скажется на распределении долей платформ в общих продажах.

```python
plats_2014 = ['year_of_release']
colors_short = {}

for plat in platforms.columns[1:]: # составляем список платформ, которые пойдут на диаграмму, и словарь цветов для них
    if platforms.loc[len(platforms) - 3:, plat].isna().sum() != 3:
        plats_2014.append(plat)
        colors_short[plat] = colors_for_platforms[plat]

platforms_since_2014 = platforms.loc[len(platforms) - 3:, plats_2014]

platforms_since_2014['total'] = platforms_since_2014[platforms_since_2014.columns[1:]].sum(axis=1)

subbar_labels = []

for col in platforms_since_2014.columns[1:-1]: # создаём список, в котором будут подписи к каждому
    container_labels_full = (                  # контейнеру в виде процента от общих продаж в данном году,
            (                                  # при этом "маленькие" контейнеры (<5%) мы оставим без подписи
                (round
                     (platforms_since_2014[col] / platforms_since_2014['total'] * 100, 2)
                .astype('str'))
            + '%')
        .to_list()
    )
    container_labels_short = [label if float(label[:-1]) > 5 else '' for label in container_labels_full]
    subbar_labels.append(container_labels_short)
    

    
(
    platforms_since_2014
        .iloc[:, : platforms_since_2014.shape[1] - 1]
        .plot(
               x='year_of_release',
               kind='bar',
               stacked=True,
               figsize=(15, 12),
               color=colors_short,
               rot=0
        )
)
plt.xlabel('год выпуска')
plt.ylabel('млн проданных копий')
plt.title('Количество проданных копий игр по платформам в 2014–2016 гг.', fontsize = 18, fontweight='bold')

for i, c in enumerate(plt.gca().containers):
    plt.gca().bar_label(
        c,
        labels=subbar_labels[i],
        label_type='center',
        color='w',
        fontsize = 16,
        fontweight='bold')
    
plt.show()
```

Судя по диаграмме выше, доминирующими платформами по количеству проданных копий в последние три года являлись PS4, XOne, 3DS, X360 и WiiU. Сокращение доли в общих продажах на протяжении всего базового периода происходит у PS3, X360 и WiiU, также скорее снижается доля 3DS. Продажи игр на платформе PS4, наоборот, серьёзно растут. Скорее рост демонстрирует и XOne. По остальным платформам особой динамики не прослеживается.

К потенциально прибыльным платформам можно отнести PS4 и XOne, плюс к этой категории тяготеет 3DS, хотя она уже, судя по всему, доживает свой век.

Построим диаграммы «ящик с усами» по глобальным продажам в разбивке по платформам. В качестве периода возьмём базовый период. На первой диаграмме будет весь диапазон значений, а на второй он будет ограничен значением в 2 млн копий.

### Связь между отзывами и продажами

Чтобы определить, есть ли связь между, с одной стороны, отзывами критиков и пользователей и, с другой стороны, общими продажами игр на платформе, мы построим две диаграммы рассеяния для самой крупной платформы в базовом периоде — PS4. По первой диаграмме можно оценить степень коррелляции продаж и отзывов критиков, а по второй — продаж и отзывов пользователей. Сразу же посчитаем коэффициент корреляции Пирсона для каждой из этих двух пар столбцов. 

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)

def draw_two_plots_with_corr(platform):    # оформим построение этих диаграмм рассеяния в виде функции,
    fig.suptitle(                          # чтобы можно было построить их и для другой платформы
        f'Связь между оценками экспертов или пользователей и продажами игр на платформе {platform}',
        y=1.01,
        fontsize=16,
        fontweight='bold')
    for fig_col, col in enumerate(['critic_score', 'user_score']):
        (
            games_since_2014
            .query('platform == @platform')
            .plot(
                kind='scatter',
                x=col,
                y='total_sales',
                color='b',
                ax=ax[fig_col]
            )
        )

        corr = 'Корреляция Пирсона: ' + str(round(
                    games_since_2014
                         .query('platform == @platform')
                         [['total_sales', col]]
                         .corr()
                         .iloc[0, 1],
                    3)
                )
        
        ax[fig_col].text(0.05, 0.9, corr, bbox=props, fontsize=14, transform=ax[fig_col].transAxes)
        ax[fig_col].set_ylabel('млн проданных копий')
        ax[fig_col].set_xlabel('значение оценки')
        ax[fig_col].set_title((f'total_sales vs. {col}'), fontsize=14)
        
draw_two_plots_with_corr('PS4')        
```

По этим диаграммам видно, что хоть какая-то линейная корреляция существует только между общими продажами и оценкой критиков (и по графику, и по значению коэффициента Пирсона видно, что корреляция слабая). Линейной корреляции между продажами и оценками пользователей нет.

Проверим эти выводы, построив такие же диаграммы рассеяния для платформы XOne:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

draw_two_plots_with_corr('XOne')
```

Получилось очень похоже на то, что мы наблюдали у PS4.

В качестве ещё одного теста сопоставим медианные продажи по платформам из базового периода с медианными оценками экспертов и пользователей:

```python
games_since_2014.pivot_table(
    index='platform',
    values=['user_score', 'critic_score', 'total_sales'],
    aggfunc='median')
```

### Портреты пользователей каждого региона

Выясним, какие платформы и жанры наиболее популярны среди жителей разных регионов. Также попробуем установить, есть ли связь между рейтингом ESRB и объёмом продаж в каждом регионе.

Для того, чтобы решить эти задачи, нам нужно будет строить круговые и столбчатые диаграммы. Первые понадобятся для определения популярных самых платформ и жанров, а вторые — для оценки влияния рейтинга ESRB на продажи (т.к. рейтинг ESRB — категориальная переменная, диаграмма рассеяния тут не подойдёт). Оформим построении круговых диаграмм в виде функции `pie_chart_for_region`, а столбчатых диаграмм — в виде функции `bar_plot_for_region`.

```python
def pie_chart_for_region(region_sales, index_col, ax, title, colors):
    top5_and_other_df = ( # собираем датафрейм с пятью наиболее популярными платформами/жанрами в регионе и их продажами
                games_since_2014
                .pivot_table(
                    index=index_col,
                    values=region_sales,
                    aggfunc='sum')
                .sort_values(
                    by=region_sales,
                    ascending=False)
                .reset_index()
               ) 

    top5_and_other_df = top5_and_other_df.append( # все остальные платформы/жанры складываем в категорию "Другие"
            {
                index_col: 'Другие',
                region_sales: top5_and_other_df.loc[5:, region_sales].sum()
            },
            ignore_index=True,
        )


    top5_and_other_df.drop(labels=list(top5_and_other_df.index)[5:-1], inplace=True)
    
    wedges_colors = [colors[key] for key in top5_and_other_df[index_col]] # генерируем список цветов, чтобы 
                                                                          # цвет одной и той же платформы не 
    top5_and_other_df.plot(                                               # отличался на разных графиках
        y=region_sales,
        kind='pie',
        ax=ax,
        autopct="%.1f%%",
        labels=None,
        colors=wedges_colors,
        textprops={
            'color': "w",
            'fontweight': 'bold',
            'fontsize': 11},
        wedgeprops={
            'linewidth': 0.5,
            'edgecolor': 'white'}
    )
    ax.legend(loc=0, labels=top5_and_other_df[index_col], fontsize=9)
    ax.set_ylabel('')
    ax.set_title(title, style='italic', fontsize=14)
    plt.tight_layout()
    
    
    
    
def bar_plot_for_region(region_sales, ax, title):
    (
        games_since_2014
         .pivot_table(
             index='rating',
             values=region_sales,
             aggfunc='sum')
         .reset_index()
         .plot(
             x='rating',
             y=region_sales,
             kind='bar',
             rot=0,
             legend=None,
             ax=ax)
    )

    ax.set_ylabel('млн проданных копий')
    ax.set_xlabel('рейтинг ESRB')
    ax.set_title(title, style='italic', fontsize=14)
    plt.tight_layout()
```

Построим теперь диаграммы для анализа долей платформ в регионах:

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(17, 5))
    
colors_for_platforms['Другие'] = '#0504aa' # зададим цвет для сегмента диаграммы "Другие"
    
pie_chart_for_region('na_sales', 'platform', ax1, 'Северная Америка', colors_for_platforms)
pie_chart_for_region('eu_sales', 'platform', ax2, 'Европа', colors_for_platforms)
pie_chart_for_region('jp_sales', 'platform', ax3, 'Япония', colors_for_platforms)

plt.suptitle('Доли платформ в региональных продажах', y=1.04, fontsize=16, fontweight='bold');
```

По диаграммам видно, что пользователи в ЕС и США в целом похожи по своим предпочтениям: в этих регионах наиболее популярны игры на платформах PS и XOne (в сумме более 60% продаж). При этом в США популярность PS4 лишь немного больше, чем XOne, а в ЕС PS4 заметно популярнее. Относительную привлекательность в США сохраняют игры на X360, а в Европе — на PC.

Япония серьёзно отличается от двух других регионов. В неё самая большая роль принадлежит платформе 3DS (около 50%), а также в ходу непопулярные или менее популярные на Западе PSV, WiiU и PS3. Популярность PS4 в Японии приличная, но и близко не такая, как в Европе и США.

Теперь посмотрим на различия между регионами с точки зрения самых популярных жанров.

```python
colors_for_genres = {
    genre : color for genre, color in zip(
        games['genre'].unique(),
        list(mcolors.XKCD_COLORS.values())[:len(games['genre'].unique()):-1])    
} # подбор цвета для каждого из жанров

colors_for_genres['Другие'] = '#0504aa'

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(17, 5))

pie_chart_for_region('na_sales', 'genre', ax1, 'Северная Америка', colors_for_genres)
pie_chart_for_region('eu_sales', 'genre', ax2, 'Европа', colors_for_genres)
pie_chart_for_region('jp_sales', 'genre', ax3, 'Япония', colors_for_genres)

plt.suptitle('Доли жанров в региональных продажах', y=1.04, fontsize=16, fontweight='bold');
```

## Статистический анализ данных

### Гипотеза о равенстве средних пользовательских оценок Xbox One и PC

В соответствии с заданием необходимо проверить две гипотезы. Начнём cо следующей гипотезы: «Средние пользовательские рейтинги платформ Xbox One и PC одинаковые».

Формализуем это утверждение, сформулировав нулевую и альтернативную гипотезы:

*H0* — средний пользовательский рейтинг игр на платформе Xbox One равен среднему пользовательскому рейтингу игр на платформе PC;

*H1* — средний пользовательский рейтинг игр на платформе Xbox One **не** равен среднему пользовательскому рейтингу игр на платформе PC.

Проверим гипотезу *H0*, используя t-тест для двух независимых генеральных совокупностей. Вывод будем делать по p-значению: если оно меньше, чем уровень значимости, то есть основания отвергнуть гипотезу *H0* и перейти к гипотезе *H1*; в противном случае гипотеза *H0* не отвергается. Можно было бы посчитать t-статистику и проверить гипотезы по ней, но с p-значением всё проще, поэтому не будем этого делать.

Установим уровень значимости в 5%.

```python
alpha = 0.05

results = st.ttest_ind(
    games_since_2014.query('platform == "XOne"')['user_score'],
    games_since_2014.query('platform == "PC"')['user_score'],
    nan_policy = 'omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Гипотеза H0 отвергается.')
else:
    print('Нет оснований отвергнуть гипотезу H0.')
```

Судя по весьма большому p-значению, средние оценки пользователей игр на платформах Xbox One и PC всё-таки **совпадают** даже при уровне значимости в 10%.

### Гипотеза о равенстве средних пользовательских оценок жанров Action и Sports

Аналогичным образом проверим вторую, такую же по своей сути гипотезу: «Средние пользовательские рейтинги жанров Action и Sports разные».

*H0* — средний пользовательский рейтинг игр жанра Action равен среднему пользовательскому рейтингу игр жанра Sports;

*H1* — средний пользовательский рейтинг игр жанра Action **не** равен среднему пользовательскому рейтингу игр жанра Sports.

Уровень значимости тот же — 5%.

```python
results = st.ttest_ind(
    games_since_2014.query('genre == "Action"')['user_score'],
    games_since_2014.query('genre == "Sports"')['user_score'],
    nan_policy = 'omit')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Гипотеза H0 отвергается.')
else:
    print('Нет оснований отвергнуть гипотезу H0.')
```

Малое p-значение говорит о том, что средние оценки пользователей игр жанров Action и Sports **не совпадают** даже при очень небольшом уровне значимости.
